```python
import matplotlib.pyplot as plt
customers = ['ABC', 'DEF', 'GHI', 'JKL', 'MNO']
customers_index = range(len(customers))
sale_amounts = [123, 90, 201, 111, 232]
```

그래프에 사용할 데이터를 만든다.

xticks함수에는 인덱스와 레이블이 모두 필요하기 때문에 index도 같이 만들었다.  

```python
fig = plt.figure()
ax1 = fig.add_subplot(1,1,1)
ax1.bar(customers_index, sale_amounts, align='center', color='darkblue')
```

그래프를 만들기 위해서는, 먼저 하나의 그림(figure)를 만들고 그 안에 하위 그래프(subplot)을 추가해야 한다. 
그림 하나에 둘 이상의 그래프를 추가하는 것이 가능하므로 만들 하위 그래프의 좌표를 지정한다. ax1은 유일한 (1행, 1열, 1개)하위 그래프임을 의미한다. 

.bar 메소드를 이용해 막대그래프를 만든다. X축은 customers_index, Y축은 sale_amounts다. align='center'는 막대가 X축 레이빌의 중앙에 위치하도록 지정하며, color는 'darkblue'는 막대의 색상을 지정한다. 

```python
fig = plt.figure()
ax1 = fig.add_subplot(1,1,1)
ax1.bar(customers_index, sale_amounts, align='center', color='darkblue')
ax1.xaxis.set_ticks_position('top')
ax1.yaxis.set_ticks_position('left')


plt.xticks(customers_index, customers, rotation=90, fontsize='small')
plt.xlabel('Customer Name')
plt.ylabel('Sale Amount')
plt.title('Sale Amount per Customer')
```

xaxis.set_ticks_position을 이용해서 X축, Y축의 눈금 위치를 지정할 수 있다. 
xticks를 이용해서 눈금 레이블을 인덱스에서 실제 이름의 인덱스로 변경한다. 
rotation=0은 눈금 레이블을 수평, 90은 수직으로 지정하며, fontsize='small'은 눈금 레이블의 크기를 줄인다. 

plt.xlabel,ylabel, tatle 을 통해 X축/Y축 레이블, 그래프 제목을 추가한다.

```python
fig = plt.figure()
ax1 = fig.add_subplot(1,1,1)
ax1.bar(customers_index, sale_amounts, align='center', color='darkblue')
ax1.xaxis.set_ticks_position('bottom')
ax1.yaxis.set_ticks_position('left')
plt.xticks(customers_index, customers, rotation=0, fontsize='small')
plt.xlabel('Customer Name')
plt.ylabel('Sale Amount')
plt.title('Sale Amount per Customer')

plt.savefig('hi.png', dpi=400)
plt.show()
```

### <font color='orange'> 히스토그램(Histogram) </font>

히스토그램은 수치형 데이터의 분포를 나타낸다. 흔히 빈도(도수), 빈도밀도(도수밀도), 확률, 확률밀도등의 분포를 그릴 때 사용한다. 

```python
import numpy as np
mu1, mu2, sigma = 100, 130, 15
x1 = mu1 + sigma*np.random.randn(10000)
x2 = mu2 + sigma*np.random.randn(10000)
fig = plt.figure()
ax1 = fig.add_subplot(1,1,1)
patches = ax1.hist(x1, bins=50, normed=False, color='darkgreen')
patches = ax1.hist(x2, bins=50, normed=False, color='orange',alpha=0.5)
ax1.xaxis.set_ticks_position('bottom')
ax1.yaxis.set_ticks_position('left')
plt.xlabel('Bins')
plt.ylabel('Number of Values in Bin')
ax1.set_title('Two Frequency Distributions')
plt.savefig('histogram.png', dip=400, bbox_inches='tight')
plt.show()
```

### <font color='orange'> 산점도(Scatter Plot) </font>

산점도는 키와 몸무게, 수요와 공급 등 두 변수 간의 관계를 표현한다. 
두 변수가 양의 상관관계인지 음의 상관관계인지를 파악할 수 있다. 
또한 산점도 위에 회귀선(regression line)을 그리면 하나의 변수 값에 따라 다른 변수값이
어떻게 변하는지 예축할 수 있다. 

회귀선: 제곱오차(squared error)를 최소화 하는 직선이다. 

```python
x = np.arange(start=1., stop=15., step=1.)
y_linear = x + 5. * np.random.randn(14)
y_quadratic = x**2 + 10*np.random.randn(14)

fn_linear = np.poly1d(np.polyfit(x,y_linear,deg=1))
fn_quadratic = np.poly1d(np.polyfit(x, y_quadratic, deg=2))

fig = plt.figure()
ax1 = fig.add_subplot(1,1,1)

ax1.plot(x, y_linear, 'bo', x, y_quadratic, 'go', x, fn_linear(x),'b-', x, fn_quadratic(x), 'g-', linewidth=2)
ax1.set_title('Scatter Plots with Best Fit Lines')

plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

## <font color='orange'> 상자그림(Box Plot) </font>

상자그림은 데이터에서 얻은 5가지 통계량을 가지고 만드는 그림이다.
- 최솟값, 최댓값, 제 1사분위수, 제 2 사분위수(중앙값), 제 3 사분위수
- 상자 맨 아래에는 제 1 사분위수, 맨 위에는 제 3 사분위수, 상자 중간에는 중앙값이 표시된다.
- 수염(whisker): 상자 위아래의 선
- 특이점, 이상점(Outlier): 상자 밖에 있는 data들


https://blog.naver.com/reisei11

```python
N = 500
normal = np.random.normal(loc=0.0, scale=1.0, size=N)
lognormal = np.random.lognormal(mean=0.0, sigma=1.0, size=N)
index_value = np.random.random_integers(low=0, high=N-1, size=N)
normal_sample = normal[index_value]
lognormal_sample = lognormal[index_value]
box_plot_data = [normal, normal_sample, lognormal, lognormal_sample]
fig = plt.figure()
ax1 = fig.add_subplot(1,1,1)
box_labels = ['normal', 'normal_sample', 'lognormal', 'lognormal_sample']
ax1.boxplot(box_plot_data, notch=False, sym='.', vert=True, whis=1.5, showmeans=True, labels=box_labels)
ax1.set_title('Box Plots: Resampling of Two Distributions')
ax1.set_xlabel('Distribution')
ax1.set_ylabel('Value')
plt.show()
```

- norch = False는 상자 중간에 홈(norch)이 파이게 하지 말고 사각형으로 그리게 한다. 
- sym은 outlier를 표기할 기호를 기본값인 + 대신 .기호로 나타낸다. 
- whis는 제 1사분위수 및 제 3 사분위수를 연장한 수염의 길이를 지정한다. 
- IQR: 사분위간 범위(interqurtile range) Q3 - Q1
- showmeans : 평균 표시 여부

### <font color='blue'> 1. Figure와 subplot </font>
- matplotlib에서 그래프는 Figure 객체 내에 존재한다. 
    - plt.figure()
    - fig = plt.figure()
    - 빈도화지 생성. 
- 도화지 만으로는 빈 객체로는 그래프를 만들 수 없으므로 add_subplot을 사용해 하나 이상의 서브플롯을 생성해야 한다. 
    - plt.figure().add_subplot(1, 1, 1)
    - fig.add_subplot(1, 1, 1)
    - ax1 = fig.add_subplot(1, 1, 1)
- plt.plot()명령으로 matplotlib은 가장 최근의 figure객체와 그 서브플롯을 그린다. 
- 서브플롯이 없다면 서브플롯 하나를 생성한다. 
    - plt.plot(randn(50).cumsum(), 'k--')

        - nrows: 서브플롯의 로우 수
        - ncols: 서브플롯의 칼럼 수
        - sharex: 모든 서브플롯이 같은 x축 눈금을 사용하도록 한다.  
        - sharey: 모든 서브플롯이 같은 y축 눈금을 사용하도록 한다. 
        - subplot_kw: add_subplot()을 사용해 각 서브플롯을 생성할 때 사용할 키워드를 담고 있는 사전
        - **fig_kw: Figure를 생성할 때 추가적으로 생성하는 키워드 인자.

- Series와 DataFrame 둘다 Plot라고 하는 Pandas의 메서드를 통해 그래프를 생성할 수 있다. 
- Plots는 기본 선그래프이다.

<h5> pandas.plot메서드 인자.</h5> 

- label: 그래프의 범례이름.
- ax: 그래프를 그릴 matplotlib의 서브플롯 객체.
- style: matplotlib에 전달할 'ko--'같은 스타일의 문자열
- alpha: 투명도 (0 ~1)
- kind: 그래프의 종류: line, bar, barh, kde
- logy: Y축에 대한 로그스케일
- use_index: 객체의 색인을 눈금 이름으로 사용할지의 여부
- rot: 눈금 이름을 로테이션(0 ~ 360)
- xticks, yticks: x축, y축으로 사용할 값
- xlim, ylim: x축, y축 한계
- grid: 축의 그리드 표시할 지 여부 

<h5> pandas의 data가 DataFrame일때 plot 메서드 인자. </h5>

- subplots: 각 DataFrame의 칼럼을 독립된 서브플롯에 그린다. 
- sharex: subplots=True면 같은 X축을 공유하고 눈금과 한계를 연결한다. 
- sharey: subplots=True면 같은 Y축을 공유한다.
- figsize: 그래프의 크기, 튜플로 지정
- title: 그래프의 제목을 문자열로 지정
- sort_columns: 칼럼을 알파벳 순서로 그린다.

### <font color='orange'> 회귀선과 일변량 히스토그램을 포함한 산점도 </font> 

```python
mean, cov = [5, 10], [(1, .5), (.5, 1)]
data = np.random.multivariate_normal(mean, cov, 200)
data_frame = pd.DataFrame(data, columns=['x', 'y'])
sns.jointplot(x='x', y='y', data=data_frame, kind='reg').set_axis_labels('x','y')
plt.suptitle('joint plot of Two variables with Bivariate and Univariate Graphs')
plt.show()
```

jointplot함수를 사용하여 회귀선이 있는 산점도와 각 변수의 히스토 그램을 만들었다.

### <font color='orange'> 부스트랩 신뢰구간을 포함한 로지스틱 회귀모형 </font> 

```python
tips['big_tip'] = (tips.tip / tips.total_bill) > .15
sns.lmplot(x='total_bill', y='big_tip', data=tips, logistic=True, y_jitter=.03).set_axis_labels('Total bill','Big Tip')
plt.title('Logistic Regression of Big Tip vs Total Bill')
plt.show()
```

lmplot함수 이용
- 이항 종속변수에 대한 로지스틱 회귀선을 만든 그림이다. 
- y_jitter인수를 사용하여 1과 0사이의 값에 지터(jitter)를 가함으로써 X축을 따라 표시되는 점의 위치를 쉽게 알 수 있다.
